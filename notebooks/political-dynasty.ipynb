# The presence of political dynasties in 2019 Thai general election

```python
import pandas as pd
constituency_df = pd.read_csv("../data/candidates.csv")
partylist_df = pd.read_csv("../data/partylist_candidates.csv")
pd.set_option('display.max_rows', 1000)  
```

```python
#merge partylist candidates dataframe with constituency candidates dataframe
candidate_df=constituency_df.append(partylist_df,ignore_index=True)
candidate_df=candidate_df.drop(columns=["Unnamed: 0","CandidateId"]) #drop unnecessary columns 
```

### ตระกูลที่ลงสมัครเยอะที่สุด
สูงสุด: พลเยี่ยม 11 คน

พลเยี่ยมเป็นนามสกุลทีี่สืบมาจากเจ้าลาวและบริวารที่อพยพมาอยู่ที่บ้านแวง อำเภอโพนทองจังหวัดร้อยเอ็ด (https://th.wikipedia.org/wiki/อำเภอโพนทอง)

พลเยี่ยมอาจจะเป็นนามสกุลที่โหลก็ได้ ดังนั้นเราลองไปดูคนที่นามสกุลเดียวกันที่อยู่ในพรรคเดียวกัน หรือทั้งพรรคเดียวกันและจังหวัดเดียวกันด้วย

Note:คนที่province_name = NaaN คือผู้สมัครparty list

### ตระกูลที่อยู่พรรคเดียวกันที่ลงสมัครเยอะที่สุด
มี734นามสกุล ที่มีผู้สมัครที่ใช้นามสกุลเดียวกันและอยู่ในพรรคเดียวกันมากกว่า1คน นับเป็นจำนวน1599คน
<br>สูงสุด: หันตุลา 8 คน
<br>ตระกูลหันตุลาจากพรรคเพื่อแผ่นดินลงสมัคร ส.ส. ในภาคอีสาน

มีผู้สมัครนามสกุลรัตนเศรษฐอยู่พรรคพลังประชารัฐ6คน ผู้สมัครนามสกุลรัตนเศรษฐหลายคนเคยเป็นอดีต ส.ส. พรรคเพื่อไทย<br>(ข่าวเกี่ยวกับการย้ายพรรคของตระกูลรัตนเศรษฐ:https://www.isranews.org/isranews-scoop/67103-rattanaset.html)


```python
family_party_count = candidate_df.groupby(by=['LastName', 'PartyName'])['LastName'].count() \
                                                                             .reset_index(name='count') \
                                                                             .sort_values(['count'], ascending=False) 
family_party_count[family_party_count["count"]>1]
```
