# Extract candidates from PDFs

```python
def extract_pdf(filename, start_col=0):
    raw = parser.from_file(filename)
    data = raw['content']
    data = re.sub(r'\n+(<.+>?)\n', '', data)
    data = re.sub(r'(\n.+2562\n?)', '', data)
    data = re.sub(r'(จังหวัด.+พรรค)\n', '', data)
    data = re.sub(r'\n\n\n', '\n', data).strip()
    data = list(map(lambda r: r.split(' ')[start_col:], filter(lambda x: x, data.split('\n'))))
    data = list(map(lambda r: r[:-1] if r[-1] == '' else r, data))
    return data
```

```python
# def build_zone_name(r):
#     return '%s-เขตเลือกตั้งที่-%d' % (r['จังหวัด'], r['เขตเลือกตั้งที่'])
    
# df_post_election_zone['zone_name'] = df_post_election_zone.apply(build_zone_name, axis=1)
def get_prefixes(province):
    if province == 'กรุงเทพมหานคร':
        return dict(area='เขต', sub_area='แขวง')
    else:
        return dict(area='อำเภอ', sub_area='ตำบล')
    
```

```python
def map_zone_to_areas(df):
    election_zones = df.groupby(['จังหวัด', 'เขตเลือกตั้งที่'])

    zones = election_zones.groups.keys()
    res = []
    for z in zones:
        province, zone_id = z[0], int(z[1])
        zone = dict(province=province, zone=zone_id, areas=[], prefixes=get_prefixes(province))
        
        area = election_zones.get_group(z).groupby('อำเภอ')
        karea = area.groups.keys()
        
        for k in karea:
            d_area = dict(area=k, interior=set(), exterior=set(), subinterior=set())
            for b in area.get_group(k).to_dict('records'):
                for ss in ['interior', 'exterior', 'subinterior']:
                    if not (b[ss] != b[ss]): # hacking way of checking nan
                        d_area[ss].add(b[ss])
#             tmp = list(filter(lambda s: 'เทศบาล' not in s, d_area['interior']))
#             d_area['subinterior'] = list(filter(lambda s: 'เทศบาล' in s, d_area['interior']))
#             d_area['interior'] = tmp 
            for ss in ['interior', 'exterior', 'subinterior']:
                d_area[ss] = list(d_area[ss])
            zone['areas'].append(d_area)
        res.append(zone)
    return res

zone_to_area_result = map_zone_to_areas(df_post_election_zone_filtered)
len(zone_to_area_result)
```

```python

def get_election_zone_from_postcode(code):
    df = df_post_election_zone_filtered[ df_post_election_zone_filtered['code'] == code ]
    set_provinces = list(set(df['จังหวัด'].values))
    if len(set_provinces) > 1:
        print('there %d provinces for code %d' % (len(set_provinces), code))

#     province = set_provinces[0]
#     print('%d (%s)' % (code, province))
    
    other_cols = list(filter(lambda x: x != 'เขตเลือกตั้งที่', df.columns))
    election_zones = df.groupby(['จังหวัด', 'เขตเลือกตั้งที่'])

    zones = election_zones.groups.keys()
    res = dict(code=int(code), zones=[])
    for z in zones:

        zone = dict_zone_to_area_result[z]
        
        res['zones'].append(zone)
    return res
# get_election_zone_from_postcode(18220)
# get_election_zone_from_postcode(12120)
# get_election_zone_from_postcode(25110)
get_election_zone_from_postcode(30450)
```

# Zone to Parties JSON

```python
zone2parties = dict()
for province_name, zone_number, party_name, can_no in  df_candidates[['province_name', 'zone_number', 'PartyName', 'CandidateNo']].values:
    key = '%s-%s' % (province_name, zone_number)
    data = (party_name, can_no)
    if key in zone2parties.keys():
        zone2parties[key].append(data)
    else:
        zone2parties[key] = [data]

for k, v in zone2parties.items():
    v = list(map(lambda x: x[0], sorted(v, key=lambda x: x[1])))
    zone2parties[k] = v
```

```python
with open('../election62/statics/zone-to-parties.json', 'w') as outfile:
    json.dump(zone2parties, outfile, ensure_ascii=False)
```
