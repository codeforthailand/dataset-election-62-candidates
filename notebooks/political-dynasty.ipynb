# The presence of political dynasties in 2019 Thai general election

```python
import pandas as pd
candidate_df = pd.read_csv("../data/candidates.csv")
pd.set_option('display.max_rows', 500) 
family_count = candidate_df.groupby(by=['LastName', 'PartyName'])['LastName'].count() \
                                                                             .reset_index(name='count') \
                                                                             .sort_values(['count'], ascending=False) 
```

### ตระกูลที่อยู่พรรคเดียวกันที่ลงสมัครเยอะที่สุด

```python
family_party_count = candidate_df.groupby(by=['LastName', 'PartyName'])['LastName'].count() \
                                                                             .reset_index(name='count') \
                                                                             .sort_values(['count'], ascending=False) 
family_party_count[family_party_count["count"]>1]
```
